# Simple Linear Regression (Regression)

## Program 1

Imported all libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

created the dataset

```python
area=np.array([1500,2000,2500,3000,3500]).reshape(-1,1) #x
price=np.array([300000,400000,500000,600000,700000])    #y
```

Create the model and train it.

```python
model=LinearRegression()
model.fit(area,price)
```

Make the precision of the price

```python
predicted_price=model.predict([[5000]])
print(predicted_price)
```

Show the Slope and Intercept

```python
print(f"Slope(Coefficient):{model.coef_[0]}",predicted_price)
print(f"Intercept:{model.intercept_}")
```

Printing the graph : predicted prices are the regression line and actual prices are the points

```python
predicted_price=model.predict(area)
plt.scatter(area,price,color='blue',label='Actual Prices')
plt.plot(area,predicted_price,color='red',label="Regression Line")
plt.xlabel("Area(Square feet)")
plt.ylabel("Prices(Dollars)")
plt.title("Simple Linear Regression: Area v/s Russia")
plt.legend()
plt.show()
```

Training the model

```python
model=LogisticRegression()
model.fit(x,y)
```

Creating figure and plotting the data point

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y,color='red',label='Data Points')
```

Plotting the regression curve

```python
x_test=np.linspace(x.min(),x.max(),1000).reshape(-1,1)
y_prob=model.predict_proba(x_test)[:,1]
```

```python
plt.scatter(x,y,color='red',label='Data Points')
plt.plot(x_test,y_prob,color='blue',linewidth=2,label='Logistic Regression curve')
plt.xlabel("Feature")
plt.ylabel('Probability')
plt.title('Logistic Regression')
plt.legend()
plt.show()
```

## Method 1

Splitting  the data into 4 different folds

```python
from sklearn.model_selection import StratifiedKFold
s_fold=StratifiedKFold(n_splits=4)
s_fold
```

```python
def score_model (model,xtrain,xtest,ytrain,ytest):
  model.fit(xtrain,ytrain)
  return model.score(xtest,ytest)
```

Uploading digits dataset

```python
from sklearn.datasets import load_digits
digits=load_digits()
dir(digits)
```

Creating x and y variables

```python
dig_data=digits.data #x
dig_target=digits.target #y
```

Importing the models

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

Creating empty lists to append the scores of each model

```python
l_score=[]
rf_score=[]
svc_score=[]
```

Splitting the data using for loop and appending te scores

```python
for train_index,test_index in s_fold.split(dig_data,dig_target):
  xtrain,xtest=digits.data[train_index],digits.data[test_index]
  ytrain,ytest=digits.target[train_index],digits.target[test_index]
  l_score.append(score_model(LogisticRegression(max_iter=500),xtrain,xtest,ytrain,ytest))
  rf_score.append(score_model(RandomForestClassifier(),xtrain,xtest,ytrain,ytest))
  svc_score.append(score_model(SVC(),xtrain,xtest,ytrain,ytest))
```

Printing the scores

```python
print(l_score)
print(rf_score)
print(svc_score)
```

# K-means algorithm

## Example1

Importing libraries and make_blobs dataset

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

Creating x and y variables and plotting the groups

```python
X,y=make_blobs(n_samples=500,n_features=2,centers=3,random_state=23)
fig=plt.figure(0)
plt.grid=True
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
k=3
clusters={}
np.random.seed(23)
for idx in range(k):
  center=2*(2*np.random.random((X.shape[1],))-1)
  points=[]
  cluster={
      'center':center,
      'points':[]
  }
  clusters[idx]=cluster
clusters
```

```python
plt.scatter(X[:,0],X[:,1])
plt.grid=True
for i in clusters:
  center=clusters[i]['center']
  plt.scatter(center[0],center[1],marker='*',color='red')
plt.show()
```

```python
def distance(p1,p2):
  return np.sqrt(np.sum((p1-p2)**2))
```

```python
def assign_clusters(X,clusters):
  for idx in range(X.shape[0]):
    dist=[]
    curr_x=X[idx]
    for i in range(k):
      dis=distance(curr_x,clusters[i]['center'])
      dist.append(dis)
    curr_cluster=np.argmin(dist)
    clusters[curr_cluster]['points'].append(curr_x)
  return clusters
```

```python
def update_clusters(X,clusters):
  for i in range(k):
    points=np.array(clusters[i]['points'])
    if points.shape[0]>0:
      new_center=points.mean(axis=0)
      clusters[i]['center']=new_center
      clusters[i]['points']=[]
  return clusters
```

```python
def pred_cluster(X,clusters):
  pred=[]
  for i in range(X.shape[0]):
    dist=[]
    for j in range(k):
      dist.append(distance(X[i],clusters[j]['center']))
    pred.append(np.argmin(dist))
  return pred
```

```python
clusters=assign_clusters(X,clusters)
clusters=update_clusters(X,clusters)
pred=pred_cluster(X,clusters)
```

```python
plt.scatter(X[:,0],X[:,1],c=pred)
for i in clusters:
  center=clusters[i]['center']
  plt.scatter(center[0],center[1],marker='*',color='red')
plt.show()
```

Creating clusters and appending inertia

```python
from sklearn.cluster import KMeans
sse=[]
k_range=range(1,len(df))
for k in k_range:
  model=KMeans(n_clusters=k)
  model.fit(df[['Age','Income($)']])
  sse.append(model.inertia_)
sse
```

Plotting an elbow graph to get the suitable number of clusters

```python
plt.xlabel('K')
plt.ylabel('Sum of r2 error')
plt.plot(k_range,sse)
```

Assigning the no.of clusters as 3

```python
km=KMeans(n_clusters=3)
```

Training the model; Creating the clusters

```python
y_predict=km.fit_predict(df[['Age','Income($)']])
df['Cluster']=y_predict
print(df.head())
print('\n',km.cluster_centers_)
```

Assigning data points to the clusters

```python
df1=df[df['Cluster']==0]
df2=df[df['Cluster']==1]
df3=df[df['Cluster']==2]
```

Scatter plotting the data points and the centroids

# Best models

```python
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
import pandas as pd
from sklearn.model_selection import GridSearchCV
from sklearn.datasets import load_iris
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
models = {
    'svm': {
        'model': svm.SVC(gamma='auto'),
        'params': {
            'C': [1, 10, 20],
            'kernel': ['rbf', 'linear']
        }
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params': {
            'n_estimators': [1, 5, 10]
        }
    },
    'logistic': {
        'model': LogisticRegression(),
        'params': {
            'C': [1, 5, 10]
        }
    }
}
```

```python
iris = load_iris()

# Store results
scores = []
```

```python
for model_name in models:
    np = models[model_name]  # Access the model and parameters
    gds = GridSearchCV(estimator=np['model'], param_grid=np['params'], cv=5, return_train_score=True)
    gds.fit(iris.data, iris.target)

    # Append the best score and parameters for the current model
    scores.append({
        'model': model_name,
        'best_score': gds.best_score_,
        'best_params': gds.best_params_
    })
```

```python
df=pd.DataFrame(scores)
df.columns=['model','best_scores','best_params']
df
```
