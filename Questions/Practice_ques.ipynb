# Cross Validation-load_iris

```python
from sklearn.datasets import load_iris
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB,MultinomialNB
```

```python
iris=load_iris()
dir(iris)
```

```python
data=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
data['target']=iris.target
data['flower_names']=data.target.apply(lambda x: iris.target_names[x])
```

```python
data
```

```python
y=data.target
x=data.drop(['target','flower_names'],axis=1)
```

```python
l_score=cross_val_score(LogisticRegression(max_iter=500),X=iris.data,y=iris.target)
s_score=cross_val_score(SVC(),X=iris.data,y=iris.target)
rf_score=cross_val_score(RandomForestClassifier(n_estimators=15),X=iris.data,y=iris.target)
dt_score=cross_val_score(DecisionTreeClassifier(),X=iris.data,y=iris.target)
gau_score=cross_val_score(GaussianNB(),X=iris.data,y=iris.target)
mul_score=cross_val_score(MultinomialNB(),X=iris.data,y=iris.target)
```

```python
print(s_score)
print(rf_score)
print(dt_score)
print(gau_score)
print(mul_score)
print(l_score)
#dividing dataset into 5 folds
```
